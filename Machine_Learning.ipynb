```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import numpy as np
import pickle
import lazypredict
```

```python
# Import data
```

```python
df = pd.read_csv('taxi.csv')
```

```python
df.head(10)
```

```python
# Get dependent and independent variables
```

```python
X = df.drop(['Numberofweeklyriders'], axis = 1)
Y = df.Numberofweeklyriders
```

```python
Y
```

```python
X
```

```python
# Split the X,Y in train and test
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.25)
```

```python
# Let's see what regression models perform well for this data:
from lazypredict.Supervised import LazyRegressor as LR
```

```python
model = LR()
```

```python
models,pred = model.fit(x_train, x_test, y_train, y_test)
```

```python
models
```

```python
# Lets explore LinearRegression
```

```python
model_lr = LinearRegression()
```

```python
model_lr
```

```python
model_lr.fit(x_train, y_train)
```

```python
model_lr.score(x_test, y_test)
```

```python
# Export model into a pickle file:
```

```python
pickle.dump( model_lr, open('Ride_Predict', 'wb'))
```
